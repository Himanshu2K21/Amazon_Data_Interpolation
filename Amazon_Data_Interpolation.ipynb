```python
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
```

```python
# Reading values from the dataset

df= pd.read_csv('amazon.csv')
df.head(10)
```

```python
# Setting index from number to product_name

df.set_index('product_name', inplace= True)
df.head(10)
```

```python
# Checking for number null values

null_count=df.isnull().sum()
print(null_count)
```

```python
# Clean 'rating_count' column: remove commas, fill NaN with 0, convert to int

df['rating_count'] = df['rating_count'].str.replace(',', '').fillna(0).astype(int)
```

```python
# Display top 5 category of products sold

# The code defines a function to extract main categories,applies it to a DataFrame, and prints the top 3 categories by frequency.

def extract_main_category(category):
    categories= category.split('|')
    return categories[0].strip()

df['category']= df['category'].apply(extract_main_category)

top_categories =df['category'].value_counts().head(3)
print(top_categories)
```

```python
# The code calculates the average discount percentage after removing '%' signs and converting to float type.

average_discount = df['discount_percentage'].str.rstrip('%').astype(float).mean()
print('Average discount % is:- ',average_discount)
```

```python
# The code processes a DataFrame column 'discounted_price': converts to string, removes currency symbols and commas, converts to numbers, and finds the top 5 expensive products.

df['discounted_price'] = df['discounted_price'].astype(str)
df['discounted_price'] = df['discounted_price'].str.replace('₹', '')
df['discounted_price'] = df['discounted_price'].str.replace(',', '')
df['discounted_price'] = pd.to_numeric(df['discounted_price'])
top_5_expensive = df.nlargest(5, 'discounted_price')
print('5 most expensive products price:- ', top_5_expensive)
```

```python
# Average rating of [roducts

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating'].mean()
```

```python
# Average review numbers for a product

df['rating_count'] = df['rating_count'].fillna(0).astype(str)
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'].astype(float).mean()
```
